````python
# Script d'entraînement RT-DETR avec données Roboflow
# À exécuter dans Google Colab

# ============================================================================
# 1. INSTALLATION DES DÉPENDANCES
# ============================================================================

# Installation d'Ultralytics (qui inclut RT-DETR)
!pip install ultralytics roboflow

# Vérification de l'installation
import ultralytics
ultralytics.checks()

# ============================================================================
# 2. IMPORTATION DES BIBLIOTHÈQUES
# ============================================================================

import os
import torch
import yaml
from roboflow import Roboflow
from ultralytics import RTDETR
import matplotlib.pyplot as plt

# Vérification du GPU
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Utilisation du device: {device}")

# ============================================================================
# 3. TÉLÉCHARGEMENT DES DONNÉES DEPUIS ROBOFLOW
# ============================================================================

# Configuration Roboflow - Votre projet de maladies de peau de tilapia
ROBOFLOW_API_KEY = "plhcLBf54yAQCtZK15IK"
WORKSPACE_NAME = "nazihs-workspace"
PROJECT_NAME = "tilapia-skin-disease-e3eov-3owpx"
VERSION_NUMBER = 1

# Initialisation de Roboflow
rf = Roboflow(api_key=ROBOFLOW_API_KEY)
project = rf.workspace(WORKSPACE_NAME).project(PROJECT_NAME)
dataset = project.version(VERSION_NUMBER).download("coco")

print(f"Dataset téléchargé dans: {dataset.location}")

# Exploration de la structure du dataset
print("\n📁 Structure du dataset:")
import os
def explore_directory(path, max_depth=2, current_depth=0):
    items = []
    if current_depth < max_depth:
        try:
            for item in sorted(os.listdir(path)):
                item_path = os.path.join(path, item)
                if os.path.isdir(item_path):
                    print("  " * current_depth + f"📁 {item}/")
                    explore_directory(item_path, max_depth, current_depth + 1)
                else:
                    print("  " * current_depth + f"📄 {item}")
        except PermissionError:
            print("  " * current_depth + "❌ Permission denied")

explore_directory(dataset.location)

# ============================================================================
# 4. PRÉPARATION DU FICHIER DE CONFIGURATION
# ============================================================================

# Conversion du format COCO vers YOLO pour RT-DETR
# Le dataset COCO doit être converti au format YOLO
print("Conversion du format COCO vers YOLO...")

# Chemin vers les fichiers COCO
coco_train_path = os.path.join(dataset.location, "train", "_annotations.coco.json")
coco_val_path = os.path.join(dataset.location, "valid", "_annotations.coco.json")

# Détection automatique de la structure des dossiers
def detect_dataset_structure(dataset_path):
    """Détecte la structure réelle du dataset et trouve les dossiers d'images"""
    structure = {}

    # Chercher les dossiers possibles
    possible_splits = ['train', 'valid', 'val', 'test']

    for split in possible_splits:
        split_path = os.path.join(dataset_path, split)
        if os.path.exists(split_path):
            # Chercher le dossier images
            images_path = os.path.join(split_path, 'images')
            if os.path.exists(images_path):
                structure[split] = f"{split}/images"
            elif any(f.lower().endswith(('.jpg', '.jpeg', '.png')) for f in os.listdir(split_path)):
                # Images directement dans le dossier split
                structure[split] = split

    return structure

# Détection de la structure
dataset_structure = detect_dataset_structure(dataset.location)
print(f"\n🔍 Structure détectée: {dataset_structure}")

# Chemin vers les fichiers COCO (adaptation selon la structure trouvée)
def find_coco_files(dataset_path):
    """Trouve les fichiers d'annotations COCO"""
    coco_files = {}

    for root, dirs, files in os.walk(dataset_path):
        for file in files:
            if file.endswith('.coco.json') or file == '_annotations.coco.json':
                # Déterminer le split depuis le chemin
                rel_path = os.path.relpath(root, dataset_path)
                if 'train' in rel_path:
                    coco_files['train'] = os.path.join(root, file)
                elif 'valid' in rel_path or 'val' in rel_path:
                    coco_files['valid'] = os.path.join(root, file)
                elif 'test' in rel_path:
                    coco_files['test'] = os.path.join(root, file)

    return coco_files

coco_files = find_coco_files(dataset.location)
print(f"📋 Fichiers COCO trouvés: {list(coco_files.keys())}")

# Extraction des classes depuis les annotations COCO
import json

def extract_classes_from_coco(coco_json_path):
    """Extrait les noms des classes depuis un fichier COCO"""
    with open(coco_json_path, 'r') as f:
        coco_data = json.load(f)

    # Extraction des catégories
    categories = coco_data.get('categories', [])
    class_names = [cat['name'] for cat in sorted(categories, key=lambda x: x['id'])]

    return class_names, len(class_names)

# Extraction des classes depuis le premier fichier COCO disponible
if coco_files:
    first_coco_file = list(coco_files.values())[0]
    class_names, num_classes = extract_classes_from_coco(first_coco_file)
else:
    print("⚠️ Aucun fichier COCO trouvé, utilisation de classes par défaut")
    class_names = ['disease']
    num_classes = 1

print(f"🏷️ Classes détectées: {class_names}")
print(f"📊 Nombre de classes: {num_classes}")

# Création du fichier data.yaml avec la structure détectée
yaml_content_parts = [f"# Configuration du dataset Tilapia Skin Disease"]
yaml_content_parts.append(f"path: {dataset.location}")

# Ajout des chemins selon la structure détectée
for split in ['train', 'val', 'test']:
    # Conversion 'valid' -> 'val' pour compatibility YOLO
    actual_split = 'valid' if split == 'val' else split

    if actual_split in dataset_structure:
        yaml_content_parts.append(f"{split}: {dataset_structure[actual_split]}")
    elif split in dataset_structure:
        yaml_content_parts.append(f"{split}: {dataset_structure[split]}")

yaml_content_parts.extend([
    "",
    f"# Classes de maladies de peau de tilapia",
    f"nc: {num_classes}",
    f"names: {class_names}"
])

data_yaml_content = "\n".join(yaml_content_parts)

# Sauvegarde du fichier data.yaml
data_yaml_path = os.path.join(dataset.location, "data.yaml")
with open(data_yaml_path, 'w') as f:
    f.write(data_yaml_content)

print(f"\n✅ Fichier data.yaml créé: {data_yaml_path}")
print("📄 Contenu du fichier data.yaml:")
print(data_yaml_content)

# Conversion COCO vers YOLO si nécessaire
def convert_coco_to_yolo(coco_json_path, output_dir):
    """Convertit les annotations COCO vers le format YOLO"""
    import json

    if not os.path.exists(coco_json_path):
        print(f"❌ Fichier COCO non trouvé: {coco_json_path}")
        return

    print(f"🔄 Conversion: {os.path.basename(coco_json_path)}")

    with open(coco_json_path, 'r') as f:
        coco_data = json.load(f)

    # Créer le dossier de sortie pour les labels
    labels_dir = os.path.join(output_dir, "labels")
    os.makedirs(labels_dir, exist_ok=True)

    # Dictionnaire pour mapper les IDs d'images
    image_id_to_filename = {img['id']: img['file_name'] for img in coco_data['images']}
    image_id_to_size = {img['id']: (img['width'], img['height']) for img in coco_data['images']}

    # Créer un fichier label pour chaque image
    annotations_by_image = {}
    for ann in coco_data['annotations']:
        image_id = ann['image_id']
        if image_id not in annotations_by_image:
            annotations_by_image[image_id] = []
        annotations_by_image[image_id].append(ann)

    converted_count = 0
    for image_id, annotations in annotations_by_image.items():
        filename = image_id_to_filename[image_id]
        width, height = image_id_to_size[image_id]

        # Nom du fichier label (même nom que l'image mais avec extension .txt)
        label_filename = os.path.splitext(filename)[0] + '.txt'
        label_path = os.path.join(labels_dir, label_filename)

        with open(label_path, 'w') as f:
            for ann in annotations:
                # Conversion bbox COCO (x, y, width, height) vers YOLO (x_center, y_center, width, height) normalisé
                x, y, w, h = ann['bbox']
                x_center = (x + w/2) / width
                y_center = (y + h/2) / height
                w_norm = w / width
                h_norm = h / height

                # Classe (COCO utilise category_id, YOLO utilise l'index de classe)
                class_id = ann['category_id'] - 1  # COCO commence à 1, YOLO à 0

                f.write(f"{class_id} {x_center:.6f} {y_center:.6f} {w_norm:.6f} {h_norm:.6f}\n")

        converted_count += 1

    print(f"✅ {converted_count} fichiers de labels convertis dans {labels_dir}")

# Conversion des annotations pour tous les splits disponibles
print("\n🔄 Conversion des annotations COCO vers YOLO...")
for split, coco_path in coco_files.items():
    split_dir = os.path.dirname(coco_path)
    convert_coco_to_yolo(coco_path, split_dir)

print("✅ Conversion COCO vers YOLO terminée!")

# Vérification finale de la structure
print("\n🔍 Vérification finale de la structure:")
for split in ['train', 'valid', 'val', 'test']:
    split_dir = os.path.join(dataset.location, split)
    if os.path.exists(split_dir):
        images_dir = os.path.join(split_dir, 'images')
        labels_dir = os.path.join(split_dir, 'labels')

        if os.path.exists(images_dir):
            img_count = len([f for f in os.listdir(images_dir) if f.lower().endswith(('.jpg', '.jpeg', '.png'))])
            print(f"📁 {split}/images: {img_count} images")

        if os.path.exists(labels_dir):
            label_count = len([f for f in os.listdir(labels_dir) if f.endswith('.txt')])
            print(f"🏷️ {split}/labels: {label_count} labels")

# Lecture et affichage de la configuration
with open(data_yaml_path, 'r') as f:
    data_config = yaml.safe_load(f)

print("Configuration du dataset:")
print(f"- Nombre de classes: {data_config['nc']}")
print(f"- Classes: {data_config['names']}")
print(f"- Chemin train: {data_config['train']}")
print(f"- Chemin val: {data_config['val']}")
if 'test' in data_config:
    print(f"- Chemin test: {data_config['test']}")

# ============================================================================
# 5. INITIALISATION DU MODÈLE RT-DETR
# ============================================================================

# Choix du modèle pré-entraîné
# Options disponibles: rtdetr-l.pt, rtdetr-x.pt
model_name = "rtdetr-l.pt"  # Modèle large, bon compromis performance/vitesse

# Chargement du modèle
model = RTDETR(model_name)
print(f"Modèle {model_name} chargé avec succès")

# ============================================================================
# 6. CONFIGURATION DE L'ENTRAÎNEMENT
# ============================================================================

# Paramètres d'entraînement
training_config = {
    'data': data_yaml_path,           # Chemin vers le fichier data.yaml
    'epochs': 170,                    # Nombre d'époques
    'batch': 16,                      # Taille du batch (ajustez selon votre GPU)
    'imgsz': 640,                     # Taille des images
    'device': device,                 # GPU ou CPU
    'project': 'tilapia_disease_detection_v2',  # Nouveau nom de projet
    'name': 'rtdetr_tilapia_170epochs',         # Nom spécifique pour 170 époques
    'save_period': 20,                # Sauvegarder toutes les 20 époques
    'patience': 85,                   # Early stopping patience (50% des époques)
    'lr0': 0.001,                     # Learning rate initial
    'lrf': 0.01,                      # Learning rate final
    'weight_decay': 0.0005,           # Weight decay
    'warmup_epochs': 3,               # Époques de warmup
    'warmup_momentum': 0.8,           # Momentum de warmup
    'box': 7.5,                       # Perte de boîte englobante
    'cls': 0.5,                       # Perte de classification
    'dfl': 1.5,                       # Perte de distribution focale
    'pose': 12.0,                     # Perte de pose (si applicable)
    'kobj': 1.0,                      # Perte d'objectness
    'label_smoothing': 0.0,           # Lissage des étiquettes
    'nbs': 64,                        # Taille nominale du batch
    'overlap_mask': True,             # Masques de chevauchement
    'mask_ratio': 4,                  # Ratio de masque
    'dropout': 0.0,                   # Dropout
    'val': True,                      # Validation pendant l'entraînement
    'plots': True,                    # Génération de graphiques
    'save': True,                     # Sauvegarder les checkpoints
    'save_txt': False,                # Sauvegarder les résultats en txt
    'save_conf': False,               # Sauvegarder les confidences
    'save_crop': False,               # Sauvegarder les crops
    'show_labels': True,              # Afficher les étiquettes
    'show_conf': True,                # Afficher les confidences
    'visualize': False,               # Visualiser les features
    'augment': True,                  # Augmentation de données
    'agnostic_nms': False,            # NMS agnostique aux classes
    'retina_masks': False,            # Masques haute résolution
    'half': False,                    # Précision half (FP16)
    'dnn': False,                     # Utiliser OpenCV DNN
    'verbose': True,                  # Mode verbeux
}

print("Configuration d'entraînement:")
for key, value in training_config.items():
    print(f"  {key}: {value}")

# ============================================================================
# 7. LANCEMENT DE L'ENTRAÎNEMENT
# ============================================================================

print("\n" + "="*50)
print("DÉBUT DE L'ENTRAÎNEMENT")
print("="*50)

# Lancement de l'entraînement
results = model.train(**training_config)

print("Entraînement terminé!")

# ============================================================================
# 8. ÉVALUATION DU MODÈLE
# ============================================================================

print("\n" + "="*50)
print("ÉVALUATION DU MODÈLE")
print("="*50)

# Évaluation sur le set de validation
metrics = model.val()

# Affichage des métriques principales
print(f"mAP50: {metrics.box.map50:.4f}")
print(f"mAP50-95: {metrics.box.map:.4f}")

# ============================================================================
# 9. TEST ET PRÉDICTIONS
# ============================================================================

print("\n" + "="*50)
print("TEST DU MODÈLE")
print("="*50)

# Recherche automatique du modèle sauvegardé
def find_model_file():
    """Trouve automatiquement le fichier de modèle disponible"""

    possible_paths = [
        "tilapia_disease_detection_v2/rtdetr_tilapia_170epochs/weights/best.pt",
        "tilapia_disease_detection_v2/rtdetr_tilapia_170epochs/weights/last.pt",
        "tilapia_disease_detection/rtdetr_tilapia_exp1/weights/best.pt",
        "tilapia_disease_detection/rtdetr_tilapia_exp1/weights/last.pt",
        "runs/detect/train/weights/best.pt",
        "runs/detect/train/weights/last.pt",
        "best.pt",
        "last.pt"
    ]

    for path in possible_paths:
        if os.path.exists(path):
            print(f"✅ Modèle trouvé: {path}")
            return path

    # Si aucun modèle trouvé, chercher dans tous les dossiers
    print("🔍 Recherche de modèles dans tous les dossiers...")
    for root, dirs, files in os.walk("."):
        for file in files:
            if file in ["best.pt", "last.pt"] and "weights" in root:
                full_path = os.path.join(root, file)
                print(f"✅ Modèle trouvé: {full_path}")
                return full_path

    return None

# Recherche du modèle
best_model_path = find_model_file()

if best_model_path is None:
    print("❌ Aucun modèle .pt trouvé!")
    print("💡 L'entraînement ne s'est peut-être pas terminé correctement.")
    print("⚠️ Vérifiez que l'entraînement a produit un fichier best.pt ou last.pt")

    # Affichage des dossiers disponibles pour diagnostic
    print("\n📁 Structure des dossiers disponibles:")
    for root, dirs, files in os.walk("."):
        if "tilapia" in root.lower() or "rtdetr" in root.lower() or "train" in root.lower():
            print(f"  📂 {root}")
            for file in files[:5]:  # Afficher seulement les 5 premiers fichiers
                print(f"    📄 {file}")

    best_model = None
else:
    # Chargement du meilleur modèle
    try:
        best_model = RTDETR(best_model_path)
        print(f"✅ Modèle chargé avec succès: {best_model_path}")
    except Exception as e:
        print(f"❌ Erreur lors du chargement du modèle: {e}")
        best_model = None

# Test sur quelques images du dataset de validation (seulement si le modèle est chargé)
if best_model is not None:

    # Recherche d'images de validation
    possible_val_paths = [
        os.path.join(dataset.location, "valid/images"),
        os.path.join(dataset.location, "val/images"),
        os.path.join(dataset.location, "test/images")
    ]

    val_images_path = None
    for path in possible_val_paths:
        if os.path.exists(path):
            val_images_path = path
            break

    if val_images_path:
        test_images = [f for f in os.listdir(val_images_path)
                      if f.lower().endswith(('.jpg', '.jpeg', '.png'))][:5]

        print(f"📁 Images de test trouvées: {len(test_images)} dans {val_images_path}")

        for img_name in test_images:
            img_path = os.path.join(val_images_path, img_name)

            try:
                # Prédiction
                results = best_model.predict(
                    source=img_path,
                    save=True,
                    save_txt=True,
                    save_conf=True,
                    project='tilapia_predictions',
                    name='disease_detection_test'
                )

                print(f"✅ Prédiction réussie pour: {img_name}")

            except Exception as e:
                print(f"❌ Erreur lors de la prédiction pour {img_name}: {e}")
    else:
        print("❌ Aucun dossier d'images de validation trouvé")
        print("📂 Dossiers disponibles dans le dataset:")
        for item in os.listdir(dataset.location):
            item_path = os.path.join(dataset.location, item)
            if os.path.isdir(item_path):
                print(f"  📁 {item}")
else:
    print("⚠️ Test ignoré car aucun modèle n'est disponible")

# ============================================================================
# 10. SAUVEGARDE ET EXPORT
# ============================================================================

print("\n" + "="*50)
print("EXPORT DU MODÈLE")
print("="*50)

# Export en différents formats (seulement si le modèle existe)
if best_model is not None:
    try:
        # Export ONNX
        best_model.export(format='onnx', dynamic=True)
        print("✓ Export ONNX réussi")

        # Export TensorRT (si disponible)
        try:
            best_model.export(format='engine', half=True)
            print("✓ Export TensorRT réussi")
        except:
            print("✗ Export TensorRT non disponible")

        # Export CoreML (si disponible)
        try:
            best_model.export(format='coreml')
            print("✓ Export CoreML réussi")
        except:
            print("✗ Export CoreML non disponible")

    except Exception as e:
        print(f"❌ Erreur lors de l'export: {e}")
else:
    print("⚠️ Export ignoré car aucun modèle n'est disponible")

# ============================================================================
# 11. AFFICHAGE DES RÉSULTATS
# ============================================================================

print("\n" + "="*50)
print("RÉSUMÉ DE L'ENTRAÎNEMENT")
print("="*50)

# Recherche automatique du dossier de résultats
def find_results_directory():
    """Trouve automatiquement le dossier contenant les résultats"""

    possible_dirs = [
        "tilapia_disease_detection_v2/rtdetr_tilapia_170epochs",
        "tilapia_disease_detection/rtdetr_tilapia_exp1",
        "runs/detect/train",
        "runs/detect/train1",
        "runs/detect/train2"
    ]

    for dir_path in possible_dirs:
        if os.path.exists(dir_path):
            print(f"📂 Dossier de résultats trouvé: {dir_path}")
            return dir_path

    # Recherche automatique
    for root, dirs, files in os.walk("."):
        if "results.png" in files or "best.pt" in files:
            print(f"📂 Dossier de résultats trouvé: {root}")
            return root

    return None

results_path = find_results_directory()

if results_path:
    # Affichage des courbes d'entraînement
    try:
        results_img = os.path.join(results_path, "results.png")
        if os.path.exists(results_img):
            from IPython.display import Image, display
            print("📈 Courbes d'entraînement:")
            display(Image(results_img))
        else:
            print("❌ Fichier results.png non trouvé")

        # Affichage de la matrice de confusion
        confusion_matrix_img = os.path.join(results_path, "confusion_matrix.png")
        if os.path.exists(confusion_matrix_img):
            print("📊 Matrice de confusion:")
            display(Image(confusion_matrix_img))
        else:
            print("ℹ️ Matrice de confusion non disponible")

    except Exception as e:
        print(f"❌ Impossible d'afficher les images: {e}")

    print(f"\n📁 Tous les résultats sont dans: {results_path}")

    # Affichage des fichiers disponibles
    print("📄 Fichiers disponibles:")
    if os.path.exists(results_path):
        for file in os.listdir(results_path):
            if file.endswith(('.png', '.pt', '.csv', '.txt')):
                print(f"  - {file}")
else:
    print("❌ Aucun dossier de résultats trouvé")
    print("💡 L'entraînement ne s'est peut-être pas terminé correctement")

# Mise à jour des chemins pour les fonctions suivantes
if results_path and os.path.exists(os.path.join(results_path, "weights/best.pt")):
    best_model_path = os.path.join(results_path, "weights/best.pt")
elif best_model_path is None:
    print("⚠️ Aucun modèle disponible pour les fonctions suivantes")

# ============================================================================
# 12. FONCTION D'INFÉRENCE SPÉCIALISÉE POUR LES MALADIES DE TILAPIA
# ============================================================================

def detect_tilapia_disease(image_path, model_path=best_model_path, conf=0.5):
    """
    Fonction spécialisée pour détecter les maladies de peau de tilapia

    Args:
        image_path (str): Chemin vers l'image de tilapia
        model_path (str): Chemin vers le modèle RT-DETR entraîné
        conf (float): Seuil de confiance pour la détection

    Returns:
        results: Résultats de la détection avec les maladies identifiées
    """
    model = RTDETR(model_path)
    results = model.predict(
        source=image_path,
        conf=conf,
        save=True,
        show_labels=True,
        show_conf=True,
        project='tilapia_disease_diagnosis',
        name='detection_results'
    )

    # Affichage des résultats spécifiques aux maladies
    for r in results:
        if len(r.boxes) > 0:
            print(f"\n🐟 Analyse de l'image: {os.path.basename(image_path)}")
            print("🦠 Maladies détectées:")
            for box in r.boxes:
                class_id = int(box.cls[0])
                confidence = float(box.conf[0])
                class_name = r.names[class_id]
                print(f"  - {class_name}: {confidence:.2%} de confiance")
        else:
            print(f"\n✅ Aucune maladie détectée sur: {os.path.basename(image_path)}")

    return results

# Exemple d'utilisation spécialisée:
# results = detect_tilapia_disease("chemin/vers/photo_tilapia.jpg", conf=0.6)

print("\n✅ Script d'entraînement RT-DETR pour détection de maladies de tilapia configuré!")
print("🔄 Configuration: 170 époques d'entraînement")
print("🐟 Modèle spécialisé dans la détection des maladies de peau de tilapia")
print("🚀 Utilisez la fonction detect_tilapia_disease() pour diagnostiquer vos poissons")
print("📊 Les résultats incluront le type de maladie et le niveau de confiance")
print("⏱️ Temps d'entraînement estimé: ~3-4 heures selon votre GPU")
````

````python
# ============================================================================
# AFFICHAGE DES RÉSULTATS RT-DETR ET VALIDATION
# Affiche: courbes, matrice de confusion, prédiction sur image
# ============================================================================

import os
import matplotlib.pyplot as plt
from IPython.display import Image, display
import pandas as pd
from ultralytics import RTDETR
import random

# ============================================================================
# 1. CONFIGURATION DES CHEMINS
# ============================================================================

# Chemin vers les résultats RT-DETR
RESULTS_DIR = "tilapia_corrected/rtdetr_170epochs_fixed"
VALIDATION_IMAGES = "/content/Tilapia-Skin-DIsease-1/valid/images"

print("📊 AFFICHAGE DES RÉSULTATS RT-DETR")
print("="*60)

# ============================================================================
# 2. FONCTION D'AFFICHAGE DES COURBES DE RÉSULTATS
# ============================================================================

def display_training_curves():
    """Affichage des courbes d'entraînement"""

    print("\n📈 COURBES D'ENTRAÎNEMENT:")

    results_png = os.path.join(RESULTS_DIR, "results.png")

    if os.path.exists(results_png):
        print("✅ Courbes mAP50, mAP50-95 et losses:")
        display(Image(results_png, width=800))
    else:
        print("❌ Fichier results.png non trouvé")
        print(f"📁 Recherche dans: {RESULTS_DIR}")

# ============================================================================
# 3. FONCTION D'AFFICHAGE DE LA MATRICE DE CONFUSION
# ============================================================================

def display_confusion_matrix():
    """Affichage de la matrice de confusion"""

    print("\n📊 MATRICE DE CONFUSION:")

    confusion_matrix_png = os.path.join(RESULTS_DIR, "confusion_matrix.png")

    if os.path.exists(confusion_matrix_png):
        print("✅ Analyse des prédictions par classe:")
        display(Image(confusion_matrix_png, width=600))
    else:
        print("❌ Fichier confusion_matrix.png non trouvé")

# ============================================================================
# 4. FONCTION D'AFFICHAGE DES MÉTRIQUES DÉTAILLÉES
# ============================================================================

def display_detailed_metrics():
    """Affichage des métriques détaillées"""

    print("\n📋 MÉTRIQUES DÉTAILLÉES:")

    results_csv = os.path.join(RESULTS_DIR, "results.csv")

    if os.path.exists(results_csv):
        # Lecture du CSV
        df = pd.read_csv(results_csv)

        # Affichage des dernières métriques
        if len(df) > 0:
            last_epoch = df.iloc[-1]

            print("🎯 RÉSULTATS FINAUX:")
            print(f"  📊 Époque finale: {int(last_epoch.get('epoch', 0))}")

            # Métriques de validation
            if 'metrics/mAP50(B)' in df.columns:
                mAP50 = last_epoch.get('metrics/mAP50(B)', 0)
                print(f"  🎯 mAP50: {mAP50:.4f} ({mAP50*100:.2f}%)")

            if 'metrics/mAP50-95(B)' in df.columns:
                mAP50_95 = last_epoch.get('metrics/mAP50-95(B)', 0)
                print(f"  🎯 mAP50-95: {mAP50_95:.4f} ({mAP50_95*100:.2f}%)")

            if 'metrics/precision(B)' in df.columns:
                precision = last_epoch.get('metrics/precision(B)', 0)
                print(f"  🎯 Précision: {precision:.4f} ({precision*100:.2f}%)")

            if 'metrics/recall(B)' in df.columns:
                recall = last_epoch.get('metrics/recall(B)', 0)
                print(f"  🎯 Rappel: {recall:.4f} ({recall*100:.2f}%)")

            # Losses
            if 'val/box_loss' in df.columns:
                val_loss = last_epoch.get('val/box_loss', 0)
                print(f"  📉 Validation Loss: {val_loss:.4f}")

            # Évolution des métriques
            print(f"\n📈 ÉVOLUTION (première → dernière époque):")
            if len(df) > 1:
                first_epoch = df.iloc[0]

                if 'metrics/mAP50(B)' in df.columns:
                    first_mAP50 = first_epoch.get('metrics/mAP50(B)', 0)
                    improvement = mAP50 - first_mAP50
                    print(f"  📊 mAP50: {first_mAP50:.4f} → {mAP50:.4f} (+{improvement:.4f})")

                if 'metrics/mAP50-95(B)' in df.columns:
                    first_mAP50_95 = first_epoch.get('metrics/mAP50-95(B)', 0)
                    improvement = mAP50_95 - first_mAP50_95
                    print(f"  📊 mAP50-95: {first_mAP50_95:.4f} → {mAP50_95:.4f} (+{improvement:.4f})")

            # Meilleure performance
            if 'metrics/mAP50(B)' in df.columns:
                best_mAP50 = df['metrics/mAP50(B)'].max()
                best_epoch_mAP50 = df['metrics/mAP50(B)'].idxmax()
                print(f"\n🏆 MEILLEURES PERFORMANCES:")
                print(f"  🥇 Meilleur mAP50: {best_mAP50:.4f} (époque {best_epoch_mAP50})")

            if 'metrics/mAP50-95(B)' in df.columns:
                best_mAP50_95 = df['metrics/mAP50-95(B)'].max()
                best_epoch_mAP50_95 = df['metrics/mAP50-95(B)'].idxmax()
                print(f"  🥇 Meilleur mAP50-95: {best_mAP50_95:.4f} (époque {best_epoch_mAP50_95})")
    else:
        print("❌ Fichier results.csv non trouvé")

# ============================================================================
# 5. FONCTION DE TEST SUR IMAGE DE VALIDATION
# ============================================================================

def test_on_validation_image():
    """Test du modèle sur une image de validation avec affichage"""

    print("\n🔍 TEST SUR IMAGE DE VALIDATION:")

    # Recherche du modèle
    best_model_path = os.path.join(RESULTS_DIR, "weights", "best.pt")

    if not os.path.exists(best_model_path):
        print(f"❌ Modèle non trouvé: {best_model_path}")
        return

    # Recherche d'images de validation
    if not os.path.exists(VALIDATION_IMAGES):
        print(f"❌ Dossier de validation non trouvé: {VALIDATION_IMAGES}")
        return

    # Sélection d'une image aléatoire
    val_images = [f for f in os.listdir(VALIDATION_IMAGES)
                  if f.lower().endswith(('.jpg', '.jpeg', '.png'))]

    if not val_images:
        print("❌ Aucune image trouvée dans le dossier de validation")
        return

    # Chargement du modèle
    try:
        model = RTDETR(best_model_path)
        print(f"✅ Modèle chargé: {best_model_path}")
    except Exception as e:
        print(f"❌ Erreur lors du chargement du modèle: {e}")
        return

    # Test sur plusieurs images
    print(f"\n🎯 TEST SUR {min(3, len(val_images))} IMAGES:")

    for i in range(min(3, len(val_images))):
        # Sélection image aléatoire
        selected_image = random.choice(val_images)
        val_images.remove(selected_image)  # Éviter les doublons

        image_path = os.path.join(VALIDATION_IMAGES, selected_image)

        print(f"\n📸 Image {i+1}: {selected_image}")

        try:
            # Prédiction
            results = model.predict(
                source=image_path,
                save=True,
                show_labels=True,
                show_conf=True,
                project='validation_test',
                name=f'test_{i+1}',
                conf=0.5
            )

            # Analyse des résultats
            if results and len(results) > 0:
                result = results[0]

                print(f"  🔍 Détections trouvées: {len(result.boxes)}")

                if len(result.boxes) > 0:
                    print(f"  🦠 MALADIES DÉTECTÉES:")
                    for j, box in enumerate(result.boxes):
                        class_id = int(box.cls[0])
                        confidence = float(box.conf[0])
                        class_name = result.names[class_id]

                        # Emoji selon la classe
                        emoji = "🦠" if "Disease" in class_name or "Streptococcus" in class_name or "TILV" in class_name else "✅"

                        print(f"    {emoji} {class_name}: {confidence:.2%} confiance")
                else:
                    print(f"  ✅ Aucune maladie détectée - Poisson probablement sain")

                # Affichage de l'image avec prédictions
                predicted_image_path = f"validation_test/test_{i+1}/{selected_image}"
                if os.path.exists(predicted_image_path):
                    print(f"  📷 Image avec prédictions:")
                    display(Image(predicted_image_path, width=400))
                else:
                    # Affichage de l'image originale
                    print(f"  📷 Image originale:")
                    display(Image(image_path, width=400))

        except Exception as e:
            print(f"  ❌ Erreur lors de la prédiction: {e}")

# ============================================================================
# 6. FONCTION D'AFFICHAGE DES COURBES SUPPLÉMENTAIRES
# ============================================================================

def display_additional_curves():
    """Affichage des courbes supplémentaires"""

    print("\n📈 COURBES SUPPLÉMENTAIRES:")

    additional_plots = [
        ("F1_curve.png", "📊 Courbe F1-Score"),
        ("PR_curve.png", "📈 Courbe Precision-Recall"),
        ("P_curve.png", "📊 Courbe de Précision"),
        ("R_curve.png", "📊 Courbe de Rappel")
    ]

    for filename, description in additional_plots:
        file_path = os.path.join(RESULTS_DIR, filename)
        if os.path.exists(file_path):
            print(f"\n{description}:")
            display(Image(file_path, width=500))

# ============================================================================
# 7. FONCTION PRINCIPALE D'AFFICHAGE
# ============================================================================

def display_all_results():
    """Fonction principale d'affichage de tous les résultats"""

    # Vérification du dossier de résultats
    if not os.path.exists(RESULTS_DIR):
        print(f"❌ Dossier de résultats non trouvé: {RESULTS_DIR}")
        print("💡 Vérifiez que l'entraînement est terminé")
        return

    # 1. Courbes d'entraînement principales
    display_training_curves()

    # 2. Matrice de confusion
    display_confusion_matrix()

    # 3. Métriques détaillées
    display_detailed_metrics()

    # 4. Test sur images de validation
    test_on_validation_image()

    # 5. Courbes supplémentaires
    display_additional_curves()

    print(f"\n🎉 AFFICHAGE TERMINÉ!")
    print(f"📋 Tous les résultats ont été affichés ci-dessus")

# ============================================================================
# 8. EXÉCUTION
# ============================================================================

# Lancement de l'affichage complet
display_all_results()

# ============================================================================
# 9. FONCTIONS UTILITAIRES BONUS
# ============================================================================

def show_model_summary():
    """Affichage du résumé du modèle"""

    best_model_path = os.path.join(RESULTS_DIR, "weights", "best.pt")

    if os.path.exists(best_model_path):
        print(f"\n🤖 RÉSUMÉ DU MODÈLE:")

        try:
            model = RTDETR(best_model_path)

            # Informations sur le modèle
            print(f"  📋 Type: RT-DETR")
            print(f"  🏷️  Classes: {len(model.names)}")
            print(f"  📝 Noms des classes:")
            for i, name in enumerate(model.names.values()):
                emoji = "🦠" if any(word in name for word in ["Disease", "Streptococcus", "TILV"]) else "✅"
                print(f"    {i}: {emoji} {name}")

            # Taille du fichier
            model_size = os.path.getsize(best_model_path) / (1024*1024)
            print(f"  💾 Taille: {model_size:.1f} MB")

        except Exception as e:
            print(f"  ❌ Erreur: {e}")

def quick_predict(image_path=None):
    """Prédiction rapide sur une image spécifiée"""

    if image_path is None:
        print("💡 Usage: quick_predict('chemin/vers/image.jpg')")
        return

    best_model_path = os.path.join(RESULTS_DIR, "weights", "best.pt")

    if os.path.exists(best_model_path) and os.path.exists(image_path):
        try:
            model = RTDETR(best_model_path)
            results = model.predict(image_path, show=True, save=True)
            print(f"✅ Prédiction effectuée sur {image_path}")
        except Exception as e:
            print(f"❌ Erreur: {e}")
    else:
        print(f"❌ Fichier non trouvé: modèle ou image")

# Affichage du résumé du modèle
show_model_summary()

print(f"\n💡 FONCTIONS DISPONIBLES:")
print(f"  📊 display_training_curves() - Afficher les courbes")
print(f"  📊 display_confusion_matrix() - Afficher la matrice")
print(f"  🔍 test_on_validation_image() - Tester sur validation")
print(f"  🚀 quick_predict('image.jpg') - Prédiction rapide")
````
