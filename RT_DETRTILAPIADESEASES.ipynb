````python
# Script d'entraÃ®nement RT-DETR avec donnÃ©es Roboflow
# Ã€ exÃ©cuter dans Google Colab

# ============================================================================
# 1. INSTALLATION DES DÃ‰PENDANCES
# ============================================================================

# Installation d'Ultralytics (qui inclut RT-DETR)
!pip install ultralytics roboflow

# VÃ©rification de l'installation
import ultralytics
ultralytics.checks()

# ============================================================================
# 2. IMPORTATION DES BIBLIOTHÃˆQUES
# ============================================================================

import os
import torch
import yaml
from roboflow import Roboflow
from ultralytics import RTDETR
import matplotlib.pyplot as plt

# VÃ©rification du GPU
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Utilisation du device: {device}")

# ============================================================================
# 3. TÃ‰LÃ‰CHARGEMENT DES DONNÃ‰ES DEPUIS ROBOFLOW
# ============================================================================

# Configuration Roboflow - Votre projet de maladies de peau de tilapia
ROBOFLOW_API_KEY = "plhcLBf54yAQCtZK15IK"
WORKSPACE_NAME = "nazihs-workspace"
PROJECT_NAME = "tilapia-skin-disease-e3eov-3owpx"
VERSION_NUMBER = 1

# Initialisation de Roboflow
rf = Roboflow(api_key=ROBOFLOW_API_KEY)
project = rf.workspace(WORKSPACE_NAME).project(PROJECT_NAME)
dataset = project.version(VERSION_NUMBER).download("coco")

print(f"Dataset tÃ©lÃ©chargÃ© dans: {dataset.location}")

# Exploration de la structure du dataset
print("\nğŸ“ Structure du dataset:")
import os
def explore_directory(path, max_depth=2, current_depth=0):
    items = []
    if current_depth < max_depth:
        try:
            for item in sorted(os.listdir(path)):
                item_path = os.path.join(path, item)
                if os.path.isdir(item_path):
                    print("  " * current_depth + f"ğŸ“ {item}/")
                    explore_directory(item_path, max_depth, current_depth + 1)
                else:
                    print("  " * current_depth + f"ğŸ“„ {item}")
        except PermissionError:
            print("  " * current_depth + "âŒ Permission denied")

explore_directory(dataset.location)

# ============================================================================
# 4. PRÃ‰PARATION DU FICHIER DE CONFIGURATION
# ============================================================================

# Conversion du format COCO vers YOLO pour RT-DETR
# Le dataset COCO doit Ãªtre converti au format YOLO
print("Conversion du format COCO vers YOLO...")

# Chemin vers les fichiers COCO
coco_train_path = os.path.join(dataset.location, "train", "_annotations.coco.json")
coco_val_path = os.path.join(dataset.location, "valid", "_annotations.coco.json")

# DÃ©tection automatique de la structure des dossiers
def detect_dataset_structure(dataset_path):
    """DÃ©tecte la structure rÃ©elle du dataset et trouve les dossiers d'images"""
    structure = {}

    # Chercher les dossiers possibles
    possible_splits = ['train', 'valid', 'val', 'test']

    for split in possible_splits:
        split_path = os.path.join(dataset_path, split)
        if os.path.exists(split_path):
            # Chercher le dossier images
            images_path = os.path.join(split_path, 'images')
            if os.path.exists(images_path):
                structure[split] = f"{split}/images"
            elif any(f.lower().endswith(('.jpg', '.jpeg', '.png')) for f in os.listdir(split_path)):
                # Images directement dans le dossier split
                structure[split] = split

    return structure

# DÃ©tection de la structure
dataset_structure = detect_dataset_structure(dataset.location)
print(f"\nğŸ” Structure dÃ©tectÃ©e: {dataset_structure}")

# Chemin vers les fichiers COCO (adaptation selon la structure trouvÃ©e)
def find_coco_files(dataset_path):
    """Trouve les fichiers d'annotations COCO"""
    coco_files = {}

    for root, dirs, files in os.walk(dataset_path):
        for file in files:
            if file.endswith('.coco.json') or file == '_annotations.coco.json':
                # DÃ©terminer le split depuis le chemin
                rel_path = os.path.relpath(root, dataset_path)
                if 'train' in rel_path:
                    coco_files['train'] = os.path.join(root, file)
                elif 'valid' in rel_path or 'val' in rel_path:
                    coco_files['valid'] = os.path.join(root, file)
                elif 'test' in rel_path:
                    coco_files['test'] = os.path.join(root, file)

    return coco_files

coco_files = find_coco_files(dataset.location)
print(f"ğŸ“‹ Fichiers COCO trouvÃ©s: {list(coco_files.keys())}")

# Extraction des classes depuis les annotations COCO
import json

def extract_classes_from_coco(coco_json_path):
    """Extrait les noms des classes depuis un fichier COCO"""
    with open(coco_json_path, 'r') as f:
        coco_data = json.load(f)

    # Extraction des catÃ©gories
    categories = coco_data.get('categories', [])
    class_names = [cat['name'] for cat in sorted(categories, key=lambda x: x['id'])]

    return class_names, len(class_names)

# Extraction des classes depuis le premier fichier COCO disponible
if coco_files:
    first_coco_file = list(coco_files.values())[0]
    class_names, num_classes = extract_classes_from_coco(first_coco_file)
else:
    print("âš ï¸ Aucun fichier COCO trouvÃ©, utilisation de classes par dÃ©faut")
    class_names = ['disease']
    num_classes = 1

print(f"ğŸ·ï¸ Classes dÃ©tectÃ©es: {class_names}")
print(f"ğŸ“Š Nombre de classes: {num_classes}")

# CrÃ©ation du fichier data.yaml avec la structure dÃ©tectÃ©e
yaml_content_parts = [f"# Configuration du dataset Tilapia Skin Disease"]
yaml_content_parts.append(f"path: {dataset.location}")

# Ajout des chemins selon la structure dÃ©tectÃ©e
for split in ['train', 'val', 'test']:
    # Conversion 'valid' -> 'val' pour compatibility YOLO
    actual_split = 'valid' if split == 'val' else split

    if actual_split in dataset_structure:
        yaml_content_parts.append(f"{split}: {dataset_structure[actual_split]}")
    elif split in dataset_structure:
        yaml_content_parts.append(f"{split}: {dataset_structure[split]}")

yaml_content_parts.extend([
    "",
    f"# Classes de maladies de peau de tilapia",
    f"nc: {num_classes}",
    f"names: {class_names}"
])

data_yaml_content = "\n".join(yaml_content_parts)

# Sauvegarde du fichier data.yaml
data_yaml_path = os.path.join(dataset.location, "data.yaml")
with open(data_yaml_path, 'w') as f:
    f.write(data_yaml_content)

print(f"\nâœ… Fichier data.yaml crÃ©Ã©: {data_yaml_path}")
print("ğŸ“„ Contenu du fichier data.yaml:")
print(data_yaml_content)

# Conversion COCO vers YOLO si nÃ©cessaire
def convert_coco_to_yolo(coco_json_path, output_dir):
    """Convertit les annotations COCO vers le format YOLO"""
    import json

    if not os.path.exists(coco_json_path):
        print(f"âŒ Fichier COCO non trouvÃ©: {coco_json_path}")
        return

    print(f"ğŸ”„ Conversion: {os.path.basename(coco_json_path)}")

    with open(coco_json_path, 'r') as f:
        coco_data = json.load(f)

    # CrÃ©er le dossier de sortie pour les labels
    labels_dir = os.path.join(output_dir, "labels")
    os.makedirs(labels_dir, exist_ok=True)

    # Dictionnaire pour mapper les IDs d'images
    image_id_to_filename = {img['id']: img['file_name'] for img in coco_data['images']}
    image_id_to_size = {img['id']: (img['width'], img['height']) for img in coco_data['images']}

    # CrÃ©er un fichier label pour chaque image
    annotations_by_image = {}
    for ann in coco_data['annotations']:
        image_id = ann['image_id']
        if image_id not in annotations_by_image:
            annotations_by_image[image_id] = []
        annotations_by_image[image_id].append(ann)

    converted_count = 0
    for image_id, annotations in annotations_by_image.items():
        filename = image_id_to_filename[image_id]
        width, height = image_id_to_size[image_id]

        # Nom du fichier label (mÃªme nom que l'image mais avec extension .txt)
        label_filename = os.path.splitext(filename)[0] + '.txt'
        label_path = os.path.join(labels_dir, label_filename)

        with open(label_path, 'w') as f:
            for ann in annotations:
                # Conversion bbox COCO (x, y, width, height) vers YOLO (x_center, y_center, width, height) normalisÃ©
                x, y, w, h = ann['bbox']
                x_center = (x + w/2) / width
                y_center = (y + h/2) / height
                w_norm = w / width
                h_norm = h / height

                # Classe (COCO utilise category_id, YOLO utilise l'index de classe)
                class_id = ann['category_id'] - 1  # COCO commence Ã  1, YOLO Ã  0

                f.write(f"{class_id} {x_center:.6f} {y_center:.6f} {w_norm:.6f} {h_norm:.6f}\n")

        converted_count += 1

    print(f"âœ… {converted_count} fichiers de labels convertis dans {labels_dir}")

# Conversion des annotations pour tous les splits disponibles
print("\nğŸ”„ Conversion des annotations COCO vers YOLO...")
for split, coco_path in coco_files.items():
    split_dir = os.path.dirname(coco_path)
    convert_coco_to_yolo(coco_path, split_dir)

print("âœ… Conversion COCO vers YOLO terminÃ©e!")

# VÃ©rification finale de la structure
print("\nğŸ” VÃ©rification finale de la structure:")
for split in ['train', 'valid', 'val', 'test']:
    split_dir = os.path.join(dataset.location, split)
    if os.path.exists(split_dir):
        images_dir = os.path.join(split_dir, 'images')
        labels_dir = os.path.join(split_dir, 'labels')

        if os.path.exists(images_dir):
            img_count = len([f for f in os.listdir(images_dir) if f.lower().endswith(('.jpg', '.jpeg', '.png'))])
            print(f"ğŸ“ {split}/images: {img_count} images")

        if os.path.exists(labels_dir):
            label_count = len([f for f in os.listdir(labels_dir) if f.endswith('.txt')])
            print(f"ğŸ·ï¸ {split}/labels: {label_count} labels")

# Lecture et affichage de la configuration
with open(data_yaml_path, 'r') as f:
    data_config = yaml.safe_load(f)

print("Configuration du dataset:")
print(f"- Nombre de classes: {data_config['nc']}")
print(f"- Classes: {data_config['names']}")
print(f"- Chemin train: {data_config['train']}")
print(f"- Chemin val: {data_config['val']}")
if 'test' in data_config:
    print(f"- Chemin test: {data_config['test']}")

# ============================================================================
# 5. INITIALISATION DU MODÃˆLE RT-DETR
# ============================================================================

# Choix du modÃ¨le prÃ©-entraÃ®nÃ©
# Options disponibles: rtdetr-l.pt, rtdetr-x.pt
model_name = "rtdetr-l.pt"  # ModÃ¨le large, bon compromis performance/vitesse

# Chargement du modÃ¨le
model = RTDETR(model_name)
print(f"ModÃ¨le {model_name} chargÃ© avec succÃ¨s")

# ============================================================================
# 6. CONFIGURATION DE L'ENTRAÃNEMENT
# ============================================================================

# ParamÃ¨tres d'entraÃ®nement
training_config = {
    'data': data_yaml_path,           # Chemin vers le fichier data.yaml
    'epochs': 170,                    # Nombre d'Ã©poques
    'batch': 16,                      # Taille du batch (ajustez selon votre GPU)
    'imgsz': 640,                     # Taille des images
    'device': device,                 # GPU ou CPU
    'project': 'tilapia_disease_detection_v2',  # Nouveau nom de projet
    'name': 'rtdetr_tilapia_170epochs',         # Nom spÃ©cifique pour 170 Ã©poques
    'save_period': 20,                # Sauvegarder toutes les 20 Ã©poques
    'patience': 85,                   # Early stopping patience (50% des Ã©poques)
    'lr0': 0.001,                     # Learning rate initial
    'lrf': 0.01,                      # Learning rate final
    'weight_decay': 0.0005,           # Weight decay
    'warmup_epochs': 3,               # Ã‰poques de warmup
    'warmup_momentum': 0.8,           # Momentum de warmup
    'box': 7.5,                       # Perte de boÃ®te englobante
    'cls': 0.5,                       # Perte de classification
    'dfl': 1.5,                       # Perte de distribution focale
    'pose': 12.0,                     # Perte de pose (si applicable)
    'kobj': 1.0,                      # Perte d'objectness
    'label_smoothing': 0.0,           # Lissage des Ã©tiquettes
    'nbs': 64,                        # Taille nominale du batch
    'overlap_mask': True,             # Masques de chevauchement
    'mask_ratio': 4,                  # Ratio de masque
    'dropout': 0.0,                   # Dropout
    'val': True,                      # Validation pendant l'entraÃ®nement
    'plots': True,                    # GÃ©nÃ©ration de graphiques
    'save': True,                     # Sauvegarder les checkpoints
    'save_txt': False,                # Sauvegarder les rÃ©sultats en txt
    'save_conf': False,               # Sauvegarder les confidences
    'save_crop': False,               # Sauvegarder les crops
    'show_labels': True,              # Afficher les Ã©tiquettes
    'show_conf': True,                # Afficher les confidences
    'visualize': False,               # Visualiser les features
    'augment': True,                  # Augmentation de donnÃ©es
    'agnostic_nms': False,            # NMS agnostique aux classes
    'retina_masks': False,            # Masques haute rÃ©solution
    'half': False,                    # PrÃ©cision half (FP16)
    'dnn': False,                     # Utiliser OpenCV DNN
    'verbose': True,                  # Mode verbeux
}

print("Configuration d'entraÃ®nement:")
for key, value in training_config.items():
    print(f"  {key}: {value}")

# ============================================================================
# 7. LANCEMENT DE L'ENTRAÃNEMENT
# ============================================================================

print("\n" + "="*50)
print("DÃ‰BUT DE L'ENTRAÃNEMENT")
print("="*50)

# Lancement de l'entraÃ®nement
results = model.train(**training_config)

print("EntraÃ®nement terminÃ©!")

# ============================================================================
# 8. Ã‰VALUATION DU MODÃˆLE
# ============================================================================

print("\n" + "="*50)
print("Ã‰VALUATION DU MODÃˆLE")
print("="*50)

# Ã‰valuation sur le set de validation
metrics = model.val()

# Affichage des mÃ©triques principales
print(f"mAP50: {metrics.box.map50:.4f}")
print(f"mAP50-95: {metrics.box.map:.4f}")

# ============================================================================
# 9. TEST ET PRÃ‰DICTIONS
# ============================================================================

print("\n" + "="*50)
print("TEST DU MODÃˆLE")
print("="*50)

# Recherche automatique du modÃ¨le sauvegardÃ©
def find_model_file():
    """Trouve automatiquement le fichier de modÃ¨le disponible"""

    possible_paths = [
        "tilapia_disease_detection_v2/rtdetr_tilapia_170epochs/weights/best.pt",
        "tilapia_disease_detection_v2/rtdetr_tilapia_170epochs/weights/last.pt",
        "tilapia_disease_detection/rtdetr_tilapia_exp1/weights/best.pt",
        "tilapia_disease_detection/rtdetr_tilapia_exp1/weights/last.pt",
        "runs/detect/train/weights/best.pt",
        "runs/detect/train/weights/last.pt",
        "best.pt",
        "last.pt"
    ]

    for path in possible_paths:
        if os.path.exists(path):
            print(f"âœ… ModÃ¨le trouvÃ©: {path}")
            return path

    # Si aucun modÃ¨le trouvÃ©, chercher dans tous les dossiers
    print("ğŸ” Recherche de modÃ¨les dans tous les dossiers...")
    for root, dirs, files in os.walk("."):
        for file in files:
            if file in ["best.pt", "last.pt"] and "weights" in root:
                full_path = os.path.join(root, file)
                print(f"âœ… ModÃ¨le trouvÃ©: {full_path}")
                return full_path

    return None

# Recherche du modÃ¨le
best_model_path = find_model_file()

if best_model_path is None:
    print("âŒ Aucun modÃ¨le .pt trouvÃ©!")
    print("ğŸ’¡ L'entraÃ®nement ne s'est peut-Ãªtre pas terminÃ© correctement.")
    print("âš ï¸ VÃ©rifiez que l'entraÃ®nement a produit un fichier best.pt ou last.pt")

    # Affichage des dossiers disponibles pour diagnostic
    print("\nğŸ“ Structure des dossiers disponibles:")
    for root, dirs, files in os.walk("."):
        if "tilapia" in root.lower() or "rtdetr" in root.lower() or "train" in root.lower():
            print(f"  ğŸ“‚ {root}")
            for file in files[:5]:  # Afficher seulement les 5 premiers fichiers
                print(f"    ğŸ“„ {file}")

    best_model = None
else:
    # Chargement du meilleur modÃ¨le
    try:
        best_model = RTDETR(best_model_path)
        print(f"âœ… ModÃ¨le chargÃ© avec succÃ¨s: {best_model_path}")
    except Exception as e:
        print(f"âŒ Erreur lors du chargement du modÃ¨le: {e}")
        best_model = None

# Test sur quelques images du dataset de validation (seulement si le modÃ¨le est chargÃ©)
if best_model is not None:

    # Recherche d'images de validation
    possible_val_paths = [
        os.path.join(dataset.location, "valid/images"),
        os.path.join(dataset.location, "val/images"),
        os.path.join(dataset.location, "test/images")
    ]

    val_images_path = None
    for path in possible_val_paths:
        if os.path.exists(path):
            val_images_path = path
            break

    if val_images_path:
        test_images = [f for f in os.listdir(val_images_path)
                      if f.lower().endswith(('.jpg', '.jpeg', '.png'))][:5]

        print(f"ğŸ“ Images de test trouvÃ©es: {len(test_images)} dans {val_images_path}")

        for img_name in test_images:
            img_path = os.path.join(val_images_path, img_name)

            try:
                # PrÃ©diction
                results = best_model.predict(
                    source=img_path,
                    save=True,
                    save_txt=True,
                    save_conf=True,
                    project='tilapia_predictions',
                    name='disease_detection_test'
                )

                print(f"âœ… PrÃ©diction rÃ©ussie pour: {img_name}")

            except Exception as e:
                print(f"âŒ Erreur lors de la prÃ©diction pour {img_name}: {e}")
    else:
        print("âŒ Aucun dossier d'images de validation trouvÃ©")
        print("ğŸ“‚ Dossiers disponibles dans le dataset:")
        for item in os.listdir(dataset.location):
            item_path = os.path.join(dataset.location, item)
            if os.path.isdir(item_path):
                print(f"  ğŸ“ {item}")
else:
    print("âš ï¸ Test ignorÃ© car aucun modÃ¨le n'est disponible")

# ============================================================================
# 10. SAUVEGARDE ET EXPORT
# ============================================================================

print("\n" + "="*50)
print("EXPORT DU MODÃˆLE")
print("="*50)

# Export en diffÃ©rents formats (seulement si le modÃ¨le existe)
if best_model is not None:
    try:
        # Export ONNX
        best_model.export(format='onnx', dynamic=True)
        print("âœ“ Export ONNX rÃ©ussi")

        # Export TensorRT (si disponible)
        try:
            best_model.export(format='engine', half=True)
            print("âœ“ Export TensorRT rÃ©ussi")
        except:
            print("âœ— Export TensorRT non disponible")

        # Export CoreML (si disponible)
        try:
            best_model.export(format='coreml')
            print("âœ“ Export CoreML rÃ©ussi")
        except:
            print("âœ— Export CoreML non disponible")

    except Exception as e:
        print(f"âŒ Erreur lors de l'export: {e}")
else:
    print("âš ï¸ Export ignorÃ© car aucun modÃ¨le n'est disponible")

# ============================================================================
# 11. AFFICHAGE DES RÃ‰SULTATS
# ============================================================================

print("\n" + "="*50)
print("RÃ‰SUMÃ‰ DE L'ENTRAÃNEMENT")
print("="*50)

# Recherche automatique du dossier de rÃ©sultats
def find_results_directory():
    """Trouve automatiquement le dossier contenant les rÃ©sultats"""

    possible_dirs = [
        "tilapia_disease_detection_v2/rtdetr_tilapia_170epochs",
        "tilapia_disease_detection/rtdetr_tilapia_exp1",
        "runs/detect/train",
        "runs/detect/train1",
        "runs/detect/train2"
    ]

    for dir_path in possible_dirs:
        if os.path.exists(dir_path):
            print(f"ğŸ“‚ Dossier de rÃ©sultats trouvÃ©: {dir_path}")
            return dir_path

    # Recherche automatique
    for root, dirs, files in os.walk("."):
        if "results.png" in files or "best.pt" in files:
            print(f"ğŸ“‚ Dossier de rÃ©sultats trouvÃ©: {root}")
            return root

    return None

results_path = find_results_directory()

if results_path:
    # Affichage des courbes d'entraÃ®nement
    try:
        results_img = os.path.join(results_path, "results.png")
        if os.path.exists(results_img):
            from IPython.display import Image, display
            print("ğŸ“ˆ Courbes d'entraÃ®nement:")
            display(Image(results_img))
        else:
            print("âŒ Fichier results.png non trouvÃ©")

        # Affichage de la matrice de confusion
        confusion_matrix_img = os.path.join(results_path, "confusion_matrix.png")
        if os.path.exists(confusion_matrix_img):
            print("ğŸ“Š Matrice de confusion:")
            display(Image(confusion_matrix_img))
        else:
            print("â„¹ï¸ Matrice de confusion non disponible")

    except Exception as e:
        print(f"âŒ Impossible d'afficher les images: {e}")

    print(f"\nğŸ“ Tous les rÃ©sultats sont dans: {results_path}")

    # Affichage des fichiers disponibles
    print("ğŸ“„ Fichiers disponibles:")
    if os.path.exists(results_path):
        for file in os.listdir(results_path):
            if file.endswith(('.png', '.pt', '.csv', '.txt')):
                print(f"  - {file}")
else:
    print("âŒ Aucun dossier de rÃ©sultats trouvÃ©")
    print("ğŸ’¡ L'entraÃ®nement ne s'est peut-Ãªtre pas terminÃ© correctement")

# Mise Ã  jour des chemins pour les fonctions suivantes
if results_path and os.path.exists(os.path.join(results_path, "weights/best.pt")):
    best_model_path = os.path.join(results_path, "weights/best.pt")
elif best_model_path is None:
    print("âš ï¸ Aucun modÃ¨le disponible pour les fonctions suivantes")

# ============================================================================
# 12. FONCTION D'INFÃ‰RENCE SPÃ‰CIALISÃ‰E POUR LES MALADIES DE TILAPIA
# ============================================================================

def detect_tilapia_disease(image_path, model_path=best_model_path, conf=0.5):
    """
    Fonction spÃ©cialisÃ©e pour dÃ©tecter les maladies de peau de tilapia

    Args:
        image_path (str): Chemin vers l'image de tilapia
        model_path (str): Chemin vers le modÃ¨le RT-DETR entraÃ®nÃ©
        conf (float): Seuil de confiance pour la dÃ©tection

    Returns:
        results: RÃ©sultats de la dÃ©tection avec les maladies identifiÃ©es
    """
    model = RTDETR(model_path)
    results = model.predict(
        source=image_path,
        conf=conf,
        save=True,
        show_labels=True,
        show_conf=True,
        project='tilapia_disease_diagnosis',
        name='detection_results'
    )

    # Affichage des rÃ©sultats spÃ©cifiques aux maladies
    for r in results:
        if len(r.boxes) > 0:
            print(f"\nğŸŸ Analyse de l'image: {os.path.basename(image_path)}")
            print("ğŸ¦  Maladies dÃ©tectÃ©es:")
            for box in r.boxes:
                class_id = int(box.cls[0])
                confidence = float(box.conf[0])
                class_name = r.names[class_id]
                print(f"  - {class_name}: {confidence:.2%} de confiance")
        else:
            print(f"\nâœ… Aucune maladie dÃ©tectÃ©e sur: {os.path.basename(image_path)}")

    return results

# Exemple d'utilisation spÃ©cialisÃ©e:
# results = detect_tilapia_disease("chemin/vers/photo_tilapia.jpg", conf=0.6)

print("\nâœ… Script d'entraÃ®nement RT-DETR pour dÃ©tection de maladies de tilapia configurÃ©!")
print("ğŸ”„ Configuration: 170 Ã©poques d'entraÃ®nement")
print("ğŸŸ ModÃ¨le spÃ©cialisÃ© dans la dÃ©tection des maladies de peau de tilapia")
print("ğŸš€ Utilisez la fonction detect_tilapia_disease() pour diagnostiquer vos poissons")
print("ğŸ“Š Les rÃ©sultats incluront le type de maladie et le niveau de confiance")
print("â±ï¸ Temps d'entraÃ®nement estimÃ©: ~3-4 heures selon votre GPU")
````

````python
# ============================================================================
# CORRECTION DU PROBLÃˆME D'IMAGES MANQUANTES
# ============================================================================

import os
import shutil

dataset_path = "/content/Tilapia-Skin-DIsease-1"

print("ğŸ”§ CORRECTION DE LA STRUCTURE DU DATASET")
print("="*50)

def fix_dataset_structure():
    """Correction de la structure du dataset"""

    for split in ['train', 'valid', 'test']:
        split_path = os.path.join(dataset_path, split)

        if os.path.exists(split_path):
            # VÃ©rifier si les images sont directement dans le dossier split
            all_files = os.listdir(split_path)
            image_files = [f for f in all_files if f.endswith(('.jpg', '.jpeg', '.png'))]

            print(f"\nğŸ“ {split}:")
            print(f"  ğŸ“„ Fichiers totaux: {len(all_files)}")
            print(f"  ğŸ–¼ï¸  Images trouvÃ©es: {len(image_files)}")

            if len(image_files) > 0:
                # CrÃ©er le dossier images s'il n'existe pas
                images_dir = os.path.join(split_path, 'images')
                os.makedirs(images_dir, exist_ok=True)

                # DÃ©placer les images
                moved_count = 0
                for img_file in image_files:
                    src = os.path.join(split_path, img_file)
                    dst = os.path.join(images_dir, img_file)

                    if not os.path.exists(dst):  # Ã‰viter les doublons
                        shutil.move(src, dst)
                        moved_count += 1

                print(f"  âœ… {moved_count} images dÃ©placÃ©es vers {split}/images/")
            else:
                print(f"  âŒ Aucune image trouvÃ©e dans {split}/")

def fix_data_yaml():
    """Correction du fichier data.yaml"""

    # Nouveau contenu corrigÃ©
    corrected_yaml = f"""# Configuration du dataset Tilapia Skin Disease
path: {dataset_path}
train: train/images
val: valid/images
test: test/images

# Classes de maladies de peau de tilapia
nc: 5
names: ['Streptococcus-pNAD', 'Bacterial Aeromonas Disease', 'Healthy Fish', 'Streptococcus', 'TILV']
"""

    yaml_path = os.path.join(dataset_path, "data.yaml")
    with open(yaml_path, 'w') as f:
        f.write(corrected_yaml)

    print(f"\nâš™ï¸ data.yaml corrigÃ©:")
    print(corrected_yaml)

def verify_fix():
    """VÃ©rification aprÃ¨s correction"""

    print(f"\nâœ… VÃ‰RIFICATION APRÃˆS CORRECTION:")

    for split in ['train', 'valid', 'test']:
        split_path = os.path.join(dataset_path, split)
        if os.path.exists(split_path):
            images_path = os.path.join(split_path, 'images')
            labels_path = os.path.join(split_path, 'labels')

            n_images = len([f for f in os.listdir(images_path) if f.endswith(('.jpg', '.png', '.jpeg'))]) if os.path.exists(images_path) else 0
            n_labels = len([f for f in os.listdir(labels_path) if f.endswith('.txt')]) if os.path.exists(labels_path) else 0

            print(f"  {split}: {n_images} images, {n_labels} labels")

            if n_images > 0 and n_labels > 0:
                print(f"    âœ… {split} OK!")
            else:
                print(f"    âŒ {split} problÃ¨me persist")

# ExÃ©cution de la correction
fix_dataset_structure()
fix_data_yaml()
verify_fix()

print(f"\nğŸ¯ CORRECTION TERMINÃ‰E!")
print(f"ğŸš€ Vous pouvez maintenant relancer l'entraÃ®nement RT-DETR!")
````

````python
# ============================================================================
# AFFICHAGE DES RÃ‰SULTATS RT-DETR ET VALIDATION
# Affiche: courbes, matrice de confusion, prÃ©diction sur image
# ============================================================================

import os
import matplotlib.pyplot as plt
from IPython.display import Image, display
import pandas as pd
from ultralytics import RTDETR
import random

# ============================================================================
# 1. CONFIGURATION DES CHEMINS
# ============================================================================

# Chemin vers les rÃ©sultats RT-DETR
RESULTS_DIR = "tilapia_corrected/rtdetr_170epochs_fixed"
VALIDATION_IMAGES = "/content/Tilapia-Skin-DIsease-1/valid/images"

print("ğŸ“Š AFFICHAGE DES RÃ‰SULTATS RT-DETR")
print("="*60)

# ============================================================================
# 2. FONCTION D'AFFICHAGE DES COURBES DE RÃ‰SULTATS
# ============================================================================

def display_training_curves():
    """Affichage des courbes d'entraÃ®nement"""

    print("\nğŸ“ˆ COURBES D'ENTRAÃNEMENT:")

    results_png = os.path.join(RESULTS_DIR, "results.png")

    if os.path.exists(results_png):
        print("âœ… Courbes mAP50, mAP50-95 et losses:")
        display(Image(results_png, width=800))
    else:
        print("âŒ Fichier results.png non trouvÃ©")
        print(f"ğŸ“ Recherche dans: {RESULTS_DIR}")

# ============================================================================
# 3. FONCTION D'AFFICHAGE DE LA MATRICE DE CONFUSION
# ============================================================================

def display_confusion_matrix():
    """Affichage de la matrice de confusion"""

    print("\nğŸ“Š MATRICE DE CONFUSION:")

    confusion_matrix_png = os.path.join(RESULTS_DIR, "confusion_matrix.png")

    if os.path.exists(confusion_matrix_png):
        print("âœ… Analyse des prÃ©dictions par classe:")
        display(Image(confusion_matrix_png, width=600))
    else:
        print("âŒ Fichier confusion_matrix.png non trouvÃ©")

# ============================================================================
# 4. FONCTION D'AFFICHAGE DES MÃ‰TRIQUES DÃ‰TAILLÃ‰ES
# ============================================================================

def display_detailed_metrics():
    """Affichage des mÃ©triques dÃ©taillÃ©es"""

    print("\nğŸ“‹ MÃ‰TRIQUES DÃ‰TAILLÃ‰ES:")

    results_csv = os.path.join(RESULTS_DIR, "results.csv")

    if os.path.exists(results_csv):
        # Lecture du CSV
        df = pd.read_csv(results_csv)

        # Affichage des derniÃ¨res mÃ©triques
        if len(df) > 0:
            last_epoch = df.iloc[-1]

            print("ğŸ¯ RÃ‰SULTATS FINAUX:")
            print(f"  ğŸ“Š Ã‰poque finale: {int(last_epoch.get('epoch', 0))}")

            # MÃ©triques de validation
            if 'metrics/mAP50(B)' in df.columns:
                mAP50 = last_epoch.get('metrics/mAP50(B)', 0)
                print(f"  ğŸ¯ mAP50: {mAP50:.4f} ({mAP50*100:.2f}%)")

            if 'metrics/mAP50-95(B)' in df.columns:
                mAP50_95 = last_epoch.get('metrics/mAP50-95(B)', 0)
                print(f"  ğŸ¯ mAP50-95: {mAP50_95:.4f} ({mAP50_95*100:.2f}%)")

            if 'metrics/precision(B)' in df.columns:
                precision = last_epoch.get('metrics/precision(B)', 0)
                print(f"  ğŸ¯ PrÃ©cision: {precision:.4f} ({precision*100:.2f}%)")

            if 'metrics/recall(B)' in df.columns:
                recall = last_epoch.get('metrics/recall(B)', 0)
                print(f"  ğŸ¯ Rappel: {recall:.4f} ({recall*100:.2f}%)")

            # Losses
            if 'val/box_loss' in df.columns:
                val_loss = last_epoch.get('val/box_loss', 0)
                print(f"  ğŸ“‰ Validation Loss: {val_loss:.4f}")

            # Ã‰volution des mÃ©triques
            print(f"\nğŸ“ˆ Ã‰VOLUTION (premiÃ¨re â†’ derniÃ¨re Ã©poque):")
            if len(df) > 1:
                first_epoch = df.iloc[0]

                if 'metrics/mAP50(B)' in df.columns:
                    first_mAP50 = first_epoch.get('metrics/mAP50(B)', 0)
                    improvement = mAP50 - first_mAP50
                    print(f"  ğŸ“Š mAP50: {first_mAP50:.4f} â†’ {mAP50:.4f} (+{improvement:.4f})")

                if 'metrics/mAP50-95(B)' in df.columns:
                    first_mAP50_95 = first_epoch.get('metrics/mAP50-95(B)', 0)
                    improvement = mAP50_95 - first_mAP50_95
                    print(f"  ğŸ“Š mAP50-95: {first_mAP50_95:.4f} â†’ {mAP50_95:.4f} (+{improvement:.4f})")

            # Meilleure performance
            if 'metrics/mAP50(B)' in df.columns:
                best_mAP50 = df['metrics/mAP50(B)'].max()
                best_epoch_mAP50 = df['metrics/mAP50(B)'].idxmax()
                print(f"\nğŸ† MEILLEURES PERFORMANCES:")
                print(f"  ğŸ¥‡ Meilleur mAP50: {best_mAP50:.4f} (Ã©poque {best_epoch_mAP50})")

            if 'metrics/mAP50-95(B)' in df.columns:
                best_mAP50_95 = df['metrics/mAP50-95(B)'].max()
                best_epoch_mAP50_95 = df['metrics/mAP50-95(B)'].idxmax()
                print(f"  ğŸ¥‡ Meilleur mAP50-95: {best_mAP50_95:.4f} (Ã©poque {best_epoch_mAP50_95})")
    else:
        print("âŒ Fichier results.csv non trouvÃ©")

# ============================================================================
# 5. FONCTION DE TEST SUR IMAGE DE VALIDATION
# ============================================================================

def test_on_validation_image():
    """Test du modÃ¨le sur une image de validation avec affichage"""

    print("\nğŸ” TEST SUR IMAGE DE VALIDATION:")

    # Recherche du modÃ¨le
    best_model_path = os.path.join(RESULTS_DIR, "weights", "best.pt")

    if not os.path.exists(best_model_path):
        print(f"âŒ ModÃ¨le non trouvÃ©: {best_model_path}")
        return

    # Recherche d'images de validation
    if not os.path.exists(VALIDATION_IMAGES):
        print(f"âŒ Dossier de validation non trouvÃ©: {VALIDATION_IMAGES}")
        return

    # SÃ©lection d'une image alÃ©atoire
    val_images = [f for f in os.listdir(VALIDATION_IMAGES)
                  if f.lower().endswith(('.jpg', '.jpeg', '.png'))]

    if not val_images:
        print("âŒ Aucune image trouvÃ©e dans le dossier de validation")
        return

    # Chargement du modÃ¨le
    try:
        model = RTDETR(best_model_path)
        print(f"âœ… ModÃ¨le chargÃ©: {best_model_path}")
    except Exception as e:
        print(f"âŒ Erreur lors du chargement du modÃ¨le: {e}")
        return

    # Test sur plusieurs images
    print(f"\nğŸ¯ TEST SUR {min(3, len(val_images))} IMAGES:")

    for i in range(min(3, len(val_images))):
        # SÃ©lection image alÃ©atoire
        selected_image = random.choice(val_images)
        val_images.remove(selected_image)  # Ã‰viter les doublons

        image_path = os.path.join(VALIDATION_IMAGES, selected_image)

        print(f"\nğŸ“¸ Image {i+1}: {selected_image}")

        try:
            # PrÃ©diction
            results = model.predict(
                source=image_path,
                save=True,
                show_labels=True,
                show_conf=True,
                project='validation_test',
                name=f'test_{i+1}',
                conf=0.5
            )

            # Analyse des rÃ©sultats
            if results and len(results) > 0:
                result = results[0]

                print(f"  ğŸ” DÃ©tections trouvÃ©es: {len(result.boxes)}")

                if len(result.boxes) > 0:
                    print(f"  ğŸ¦  MALADIES DÃ‰TECTÃ‰ES:")
                    for j, box in enumerate(result.boxes):
                        class_id = int(box.cls[0])
                        confidence = float(box.conf[0])
                        class_name = result.names[class_id]

                        # Emoji selon la classe
                        emoji = "ğŸ¦ " if "Disease" in class_name or "Streptococcus" in class_name or "TILV" in class_name else "âœ…"

                        print(f"    {emoji} {class_name}: {confidence:.2%} confiance")
                else:
                    print(f"  âœ… Aucune maladie dÃ©tectÃ©e - Poisson probablement sain")

                # Affichage de l'image avec prÃ©dictions
                predicted_image_path = f"validation_test/test_{i+1}/{selected_image}"
                if os.path.exists(predicted_image_path):
                    print(f"  ğŸ“· Image avec prÃ©dictions:")
                    display(Image(predicted_image_path, width=400))
                else:
                    # Affichage de l'image originale
                    print(f"  ğŸ“· Image originale:")
                    display(Image(image_path, width=400))

        except Exception as e:
            print(f"  âŒ Erreur lors de la prÃ©diction: {e}")

# ============================================================================
# 6. FONCTION D'AFFICHAGE DES COURBES SUPPLÃ‰MENTAIRES
# ============================================================================

def display_additional_curves():
    """Affichage des courbes supplÃ©mentaires"""

    print("\nğŸ“ˆ COURBES SUPPLÃ‰MENTAIRES:")

    additional_plots = [
        ("F1_curve.png", "ğŸ“Š Courbe F1-Score"),
        ("PR_curve.png", "ğŸ“ˆ Courbe Precision-Recall"),
        ("P_curve.png", "ğŸ“Š Courbe de PrÃ©cision"),
        ("R_curve.png", "ğŸ“Š Courbe de Rappel")
    ]

    for filename, description in additional_plots:
        file_path = os.path.join(RESULTS_DIR, filename)
        if os.path.exists(file_path):
            print(f"\n{description}:")
            display(Image(file_path, width=500))

# ============================================================================
# 7. FONCTION PRINCIPALE D'AFFICHAGE
# ============================================================================

def display_all_results():
    """Fonction principale d'affichage de tous les rÃ©sultats"""

    # VÃ©rification du dossier de rÃ©sultats
    if not os.path.exists(RESULTS_DIR):
        print(f"âŒ Dossier de rÃ©sultats non trouvÃ©: {RESULTS_DIR}")
        print("ğŸ’¡ VÃ©rifiez que l'entraÃ®nement est terminÃ©")
        return

    # 1. Courbes d'entraÃ®nement principales
    display_training_curves()

    # 2. Matrice de confusion
    display_confusion_matrix()

    # 3. MÃ©triques dÃ©taillÃ©es
    display_detailed_metrics()

    # 4. Test sur images de validation
    test_on_validation_image()

    # 5. Courbes supplÃ©mentaires
    display_additional_curves()

    print(f"\nğŸ‰ AFFICHAGE TERMINÃ‰!")
    print(f"ğŸ“‹ Tous les rÃ©sultats ont Ã©tÃ© affichÃ©s ci-dessus")

# ============================================================================
# 8. EXÃ‰CUTION
# ============================================================================

# Lancement de l'affichage complet
display_all_results()

# ============================================================================
# 9. FONCTIONS UTILITAIRES BONUS
# ============================================================================

def show_model_summary():
    """Affichage du rÃ©sumÃ© du modÃ¨le"""

    best_model_path = os.path.join(RESULTS_DIR, "weights", "best.pt")

    if os.path.exists(best_model_path):
        print(f"\nğŸ¤– RÃ‰SUMÃ‰ DU MODÃˆLE:")

        try:
            model = RTDETR(best_model_path)

            # Informations sur le modÃ¨le
            print(f"  ğŸ“‹ Type: RT-DETR")
            print(f"  ğŸ·ï¸  Classes: {len(model.names)}")
            print(f"  ğŸ“ Noms des classes:")
            for i, name in enumerate(model.names.values()):
                emoji = "ğŸ¦ " if any(word in name for word in ["Disease", "Streptococcus", "TILV"]) else "âœ…"
                print(f"    {i}: {emoji} {name}")

            # Taille du fichier
            model_size = os.path.getsize(best_model_path) / (1024*1024)
            print(f"  ğŸ’¾ Taille: {model_size:.1f} MB")

        except Exception as e:
            print(f"  âŒ Erreur: {e}")

def quick_predict(image_path=None):
    """PrÃ©diction rapide sur une image spÃ©cifiÃ©e"""

    if image_path is None:
        print("ğŸ’¡ Usage: quick_predict('chemin/vers/image.jpg')")
        return

    best_model_path = os.path.join(RESULTS_DIR, "weights", "best.pt")

    if os.path.exists(best_model_path) and os.path.exists(image_path):
        try:
            model = RTDETR(best_model_path)
            results = model.predict(image_path, show=True, save=True)
            print(f"âœ… PrÃ©diction effectuÃ©e sur {image_path}")
        except Exception as e:
            print(f"âŒ Erreur: {e}")
    else:
        print(f"âŒ Fichier non trouvÃ©: modÃ¨le ou image")

# Affichage du rÃ©sumÃ© du modÃ¨le
show_model_summary()

print(f"\nğŸ’¡ FONCTIONS DISPONIBLES:")
print(f"  ğŸ“Š display_training_curves() - Afficher les courbes")
print(f"  ğŸ“Š display_confusion_matrix() - Afficher la matrice")
print(f"  ğŸ” test_on_validation_image() - Tester sur validation")
print(f"  ğŸš€ quick_predict('image.jpg') - PrÃ©diction rapide")
````
