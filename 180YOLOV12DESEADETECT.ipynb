```python
# YOLOv12n Training Script for Google Colab
# Tilapia Skin Disease Detection

import os
import shutil
from pathlib import Path
import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import Image, display

# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Create directory for saving results
results_dir = '/content/drive/MyDrive/YOLOv12_Results'
os.makedirs(results_dir, exist_ok=True)
print(f"Results will be saved to: {results_dir}")

# Install required packages
!pip install ultralytics roboflow

# Download dataset from Roboflow
!pip install roboflow
from roboflow import Roboflow
import glob

rf = Roboflow(api_key="plhcLBf54yAQCtZK15IK")
project = rf.workspace("nazihs-workspace").project("tilapia-skin-disease-e3eov-vn5iq")
version = project.version(6)
dataset = version.download("yolov12")

# Import YOLO
from ultralytics import YOLO
import torch

# Check GPU availability
print(f"CUDA available: {torch.cuda.is_available()}")
print(f"GPU device: {torch.cuda.get_device_name(0) if torch.cuda.is_available() else 'CPU'}")

# Initialize YOLOv12n model
model = YOLO('yolo12n.pt')  # Download pretrained YOLOv12n

# Find the correct dataset path and data.yaml file
print("Looking for dataset directory...")
possible_paths = [
    f"/content/{dataset.name}-{dataset.version}",
    f"/content/Tilapia-Skin-DIsease-{dataset.version}",
    f"/content/tilapia-skin-disease-e3eov-vn5iq-{dataset.version}"
]

# Search for all directories that might contain the dataset
all_dirs = glob.glob("/content/*/")
dataset_dirs = [d for d in all_dirs if any(keyword in d.lower() for keyword in ['tilapia', 'skin', 'disease'])]

print("Found directories:")
for d in all_dirs:
    print(f"  {d}")

# Find data.yaml file
data_yaml_files = glob.glob("/content/**/data.yaml", recursive=True)
print(f"\nFound data.yaml files: {data_yaml_files}")

if data_yaml_files:
    data_yaml = data_yaml_files[0]
    dataset_path = os.path.dirname(data_yaml)
else:
    # Try to find the dataset directory manually
    if dataset_dirs:
        dataset_path = dataset_dirs[0].rstrip('/')
        data_yaml = f"{dataset_path}/data.yaml"
    else:
        # Fallback: use the expected path but check if it exists
        dataset_path = f"/content/{dataset.name}-{dataset.version}"
        data_yaml = f"{dataset_path}/data.yaml"

print(f"Dataset path: {dataset_path}")
print(f"Data config: {data_yaml}")

# Verify the data.yaml file exists
if not os.path.exists(data_yaml):
    print(f"ERROR: data.yaml not found at {data_yaml}")
    print("Available files in dataset directory:")
    if os.path.exists(dataset_path):
        for file in os.listdir(dataset_path):
            print(f"  {file}")
    else:
        print(f"Dataset directory {dataset_path} does not exist!")
        print("Available directories in /content/:")
        for item in os.listdir("/content/"):
            if os.path.isdir(f"/content/{item}"):
                print(f"  {item}")

    # Try to find any YAML file in the dataset directory
    yaml_files = glob.glob(f"{dataset_path}/*.yaml") if os.path.exists(dataset_path) else []
    if yaml_files:
        print(f"Found YAML files: {yaml_files}")
        data_yaml = yaml_files[0]
        print(f"Using: {data_yaml}")
    else:
        raise FileNotFoundError(f"No data.yaml file found. Please check the dataset download.")
else:
    print("✓ data.yaml file found successfully!")

# Training parameters
epochs = 200
img_size = 640
batch_size = 16
patience = 50

# Train the model
print("Starting training...")
results = model.train(
    data=data_yaml,
    epochs=epochs,
    imgsz=img_size,
    batch=batch_size,
    patience=patience,
    save=True,
    save_period=50,  # Save checkpoint every 50 epochs
    cache=True,
    device=0 if torch.cuda.is_available() else 'cpu',
    workers=2,
    project='/content/runs/detect',
    name='yolov12n_tilapia_disease',
    exist_ok=True,
    verbose=True
)

# Get the training run directory
run_dir = results.save_dir
print(f"Training completed! Results saved in: {run_dir}")

# Display key metrics
def display_metrics():
    """Display important training metrics"""
    results_csv = f"{run_dir}/results.csv"

    if os.path.exists(results_csv):
        df = pd.read_csv(results_csv)
        df.columns = df.columns.str.strip()  # Remove any whitespace

        print("=== TRAINING SUMMARY ===")
        print(f"Total Epochs: {len(df)}")

        # Find best epoch
        if 'metrics/mAP50(B)' in df.columns:
            best_epoch = df['metrics/mAP50(B)'].idxmax()
            best_map50 = df.loc[best_epoch, 'metrics/mAP50(B)']
            best_map50_95 = df.loc[best_epoch, 'metrics/mAP50-95(B)'] if 'metrics/mAP50-95(B)' in df.columns else 'N/A'
        else:
            # Try alternative column names
            map_cols = [col for col in df.columns if 'mAP' in col and '50' in col]
            if map_cols:
                best_epoch = df[map_cols[0]].idxmax()
                best_map50 = df.loc[best_epoch, map_cols[0]]
                best_map50_95 = 'N/A'
            else:
                best_epoch, best_map50, best_map50_95 = 'N/A', 'N/A', 'N/A'

        print(f"Best Epoch: {best_epoch}")
        print(f"Best mAP@0.5: {best_map50}")
        print(f"Best mAP@0.5:0.95: {best_map50_95}")

        # Final metrics
        final_metrics = df.iloc[-1]
        print("\n=== FINAL METRICS ===")
        for col in df.columns:
            if any(metric in col.lower() for metric in ['map', 'precision', 'recall', 'loss']):
                print(f"{col}: {final_metrics[col]:.4f}")

        return df
    else:
        print("Results CSV not found!")
        return None

# Display metrics
metrics_df = display_metrics()

# Validate the model
print("\nRunning validation...")
val_results = model.val(data=data_yaml, imgsz=img_size, batch=batch_size)

print("\n=== VALIDATION METRICS ===")
print(f"mAP@0.5: {val_results.box.map50:.4f}")
print(f"mAP@0.5:0.95: {val_results.box.map:.4f}")
print(f"Precision: {val_results.box.mp:.4f}")
print(f"Recall: {val_results.box.mr:.4f}")

# Function to copy files to Google Drive
def save_to_drive():
    """Save all important files to Google Drive"""
    print("\nSaving results to Google Drive...")

    # Files to save
    files_to_save = {
        'best.pt': 'Best model weights',
        'last.pt': 'Last epoch weights',
        'results.csv': 'Training metrics CSV',
        'results.png': 'Training curves plot',
        'confusion_matrix.png': 'Confusion matrix',
        'confusion_matrix_normalized.png': 'Normalized confusion matrix',
        'F1_curve.png': 'F1 score curve',
        'P_curve.png': 'Precision curve',
        'R_curve.png': 'Recall curve',
        'PR_curve.png': 'Precision-Recall curve',
        'val_batch0_labels.jpg': 'Validation batch labels',
        'val_batch0_pred.jpg': 'Validation batch predictions'
    }

    saved_files = []

    for filename, description in files_to_save.items():
        source_path = f"{run_dir}/{filename}"
        if os.path.exists(source_path):
            dest_path = f"{results_dir}/{filename}"
            shutil.copy2(source_path, dest_path)
            saved_files.append(f"✓ {description}: {filename}")
            print(f"Saved: {filename}")
        else:
            print(f"Not found: {filename}")

    # Save training arguments
    args_file = f"{results_dir}/training_args.txt"
    with open(args_file, 'w') as f:
        f.write("YOLOv12n Training Configuration\n")
        f.write("=" * 40 + "\n")
        f.write(f"Model: YOLOv12n\n")
        f.write(f"Dataset: Tilapia Skin Disease\n")
        f.write(f"Epochs: {epochs}\n")
        f.write(f"Image Size: {img_size}\n")
        f.write(f"Batch Size: {batch_size}\n")
        f.write(f"Patience: {patience}\n")
        f.write(f"Device: {'GPU' if torch.cuda.is_available() else 'CPU'}\n")
        f.write("\nSaved Files:\n")
        for file_info in saved_files:
            f.write(f"{file_info}\n")

    print(f"\nAll results saved to: {results_dir}")
    return saved_files

# Save results to Drive
saved_files = save_to_drive()

# Display confusion matrices if available
def display_confusion_matrices():
    """Display confusion matrices"""
    cm_files = [
        f"{results_dir}/confusion_matrix.png",
        f"{results_dir}/confusion_matrix_normalized.png"
    ]

    for cm_file in cm_files:
        if os.path.exists(cm_file):
            print(f"\nDisplaying: {os.path.basename(cm_file)}")
            display(Image(cm_file))
        else:
            print(f"Confusion matrix not found: {cm_file}")

# Display results
display_confusion_matrices()

# Display training curves
results_plot = f"{results_dir}/results.png"
if os.path.exists(results_plot):
    print("\nTraining Results:")
    display(Image(results_plot))

# Summary of saved files
print("\n" + "="*50)
print("TRAINING COMPLETED SUCCESSFULLY!")
print("="*50)
print(f"Training run directory: {run_dir}")
print(f"Google Drive backup: {results_dir}")
print("\nSaved files:")
for file_info in saved_files:
    print(f"  {file_info}")

print(f"\nBest model weights: {results_dir}/best.pt")
print("You can download these files from your Google Drive!")

# Optional: Test the trained model on a sample image
def test_sample_prediction():
    """Test the trained model on a sample image"""
    # Load the best model
    best_model = YOLO(f"{results_dir}/best.pt")

    # Find a test image from the dataset
    test_images_dir = f"{dataset_path}/test/images"
    if os.path.exists(test_images_dir):
        test_images = [f for f in os.listdir(test_images_dir) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
        if test_images:
            test_img = f"{test_images_dir}/{test_images[0]}"
            print(f"\nTesting on sample image: {test_images[0]}")

            # Run prediction
            results = best_model(test_img)

            # Save prediction result
            pred_img = f"{results_dir}/sample_prediction.jpg"
            results[0].save(pred_img)
            print(f"Sample prediction saved: {pred_img}")

            # Display the prediction
            display(Image(pred_img))
        else:
            print("No test images found for sample prediction")
    else:
        print("Test images directory not found")

# Run sample prediction
test_sample_prediction()

print("\n🎉 Training pipeline completed! Check your Google Drive for all results.")
```
